1. Difference between np.array(), np.asarray(), and np.asanyarray()
np.array():

This function creates a new NumPy array, even if the input is already a NumPy array.
It always makes a copy of the data unless specified otherwise.
Example: np.array([1, 2, 3]) will create a new array, even if the input is already an array.
np.asarray():

It converts the input to a NumPy array, but if the input is already an array, it does not make a copy.
It is more memory-efficient when you don't need a copy of an existing array.
Example: np.asarray([1, 2, 3]) will create a new array only if the input is not already an array.
np.asanyarray():

Similar to np.asarray(), but it allows for the creation of subclass arrays like matrix or masked_array.
It avoids converting subclass arrays into plain ndarray if possible.
Example: np.asanyarray(matrix) will return a matrix rather than converting it into an ndarray.

2. Difference between Deep Copy and Shallow Copy
Shallow Copy:

A shallow copy creates a new object, but it copies the references to the objects within the original.
This means changes to nested/mutable objects (like lists inside lists) in the copied object will reflect in the original object.
Example: When using copy.copy(), the top-level structure is copied, but the inner objects (like lists or arrays) are still linked to the original.
Deep Copy:

A deep copy creates a completely new object, including all the objects within it.
Changes to any nested or inner objects will not affect the original object.
Example: When using copy.deepcopy(), a fully independent copy is made, including all inner objects.

```python
#Answer-7

import pandas as pd

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Phone': ['(123) 456-7890', '987-654-3210', '555.123.4567'],
    'Age': [25, 30, 35]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Clean the 'Phone' column: remove non-numeric characters
df['Phone'] = df['Phone'].str.replace(r'\D', '', regex=True)

# Convert the 'Phone' column to a numeric data type
df['Phone'] = pd.to_numeric(df['Phone'], errors='coerce')

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)

# Display the DataFrame's attributes and data types
print("\nDataFrame attributes and data types:")
print(df.dtypes)
```

```python
#Answer-14

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a date range
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')

# Step 2: Generate random temperature and humidity data
temperature = np.random.uniform(low=15, high=30, size=100)  # Temperature between 15 and 30 degrees Celsius
humidity = np.random.uniform(low=30, high=100, size=100)    # Humidity between 30% and 100%

# Step 3: Create a DataFrame
df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Step 4: Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Step 5: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Step 6: Show the plot
plt.show()
```

```python
#Answer-15

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean = 0, stddev = 1

# Step 2: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Step 3: Overlay a line plot representing the normal distribution's PDF
# Calculate the PDF
xmin, xmax = plt.xlim()  # Get the range of x-values for the plot
x = np.linspace(xmin, xmax, 100)  # Generate 100 points in the x range
p = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF for a normal distribution with mean 0 and stddev 1

plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Step 4: Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display legend
plt.legend()

# Step 6: Show the plot
plt.show()
```

```python
#Answer-18

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

# Step 1: Set up output in the notebook
output_notebook()

# Step 2: Generate random categorical data
categories = [f'Category {i}' for i in range(10)]  # Create 10 categories
values = np.random.randint(1, 100, size=len(categories))  # Generate random values for each category

# Step 3: Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Step 4: Create a bar chart
# Changed 'plot_height' to 'height'
p = figure(x_range=categories, height=400, title='Random Categorical Bar Chart',
           toolbar_location=None, tools='')

# Step 5: Add bars with colors based on values
p.vbar(x='categories', top='values', width=0.9, source=source, line_color='white',
       fill_color='blue')

# Step 6: Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Step 7: Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'
p.xgrid.grid_line_color = None
p.y_range.start = 0  # Start y-axis from 0

# Step 8: Show the plot
show(p)
```
